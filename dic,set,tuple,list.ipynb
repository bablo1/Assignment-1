# dictionary

```python
car = {"key":"toyota,honda","color":["red","blue"]}
print(car["color"][1])
print(car["key"][::-1])
print(car["key"][0:5])
print(car["color"][1][2])
type(car)
```

# set

```python
city = {"faisalabad","khurianwala","faisalabad city","jaranwala"}
print(city)
city.add("faisalabad sadar")
print(city)
type(city)
```

# Tuple

```python
friend = ("abbas","usama","dr,usama","saad")
print(friend[::1])

print(friend[0])
type(friend)
```

# List

```python
name = ["usama","dr,usama","saad","abbas"]
print(name)
name.append("adnan") #save the option in last index
print(name)
name.insert(0,"babr") #save the option in mear index
print(name)
name.remove("abbas")
print(name)
type(name)
```
